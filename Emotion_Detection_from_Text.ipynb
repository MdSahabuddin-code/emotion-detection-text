```python
import pandas as pd
```

```python
df=pd.read_csv('test(in).csv')
```

```python
df.head()
```

```python
#tokenizing every word
total_tokens = []
unique_tokens = []
i=0;
while i< len(df):
    a=str(df['text'][i]).split(" ")
    unique_tokens = set(unique_tokens).union(set(a))
    total_tokens +=a
    i +=1
print(len(total_tokens))
print(len(unique_tokens))
print(type(total_tokens))
print(type(unique_tokens))
```

```python
#term count
word_count_dict = {}

for word in unique_tokens:
    key = word
    value = total_tokens.count(word)
    word_count_dict[key] = value
print(word_count_dict)
```

```python
#eleminating the words of word count more than 30
final_tokens = []
print('removed words: ')
for word in unique_tokens:
    if(word_count_dict[word]<30):
        final_tokens.append(word)
    else:
        print(word)
#print(final_tokens)
print('len of final tokens: ',len(final_tokens))
```

```python
# Eliminating stop words
print(len(final_tokens))

# Read stopwords file as one-column dataframe
stop_words_df = pd.read_csv('stopwords-bn.txt', header=None, names=['word'])

stop_words = []
final_tokens_1 = []

# Collect stopwords
i = 0
while i < len(stop_words_df):
    stop_words.append(stop_words_df['word'][i])
    i += 1

print(stop_words)

# Filter tokens
i = 0
while i < len(final_tokens):
    if stop_words.count(final_tokens[i]) == 0:  # check if not a stopword
        final_tokens_1.append(final_tokens[i])
    i += 1

print(len(final_tokens_1))
print(len(final_tokens))
```

```python
#breaking down into positive, negative and neutral parts
tmp_positive_tokens = []
tmp_negative_tokens = []
tmp_neutral_tokens = []
i=0
while i < len(df):
    if(df['sentiment'][i]=='positive'):
        tmp_positive_tokens += (df['text'][i].split())
    elif(df['sentiment'][i]=='negative'):
        tmp_negative_tokens += (df['text'][i].split())
    elif(df['sentiment'][i]=='neutral'):
        tmp_neutral_tokens += (df['text'][i].split())
    i += 1
```

```python
#cutting off or filtering off the unnecessary words
positive_tokens = []
print(len(final_tokens))
print('____________')
negative_tokens = []
neutral_tokens = []
for word in tmp_positive_tokens:
    if(final_tokens_1.count(word) and (positive_tokens.count(word)==0)):
        positive_tokens.append(word)
for word in tmp_negative_tokens:
    if(final_tokens_1.count(word) and (negative_tokens.count(word)==0)):
        negative_tokens.append(word)
for word in tmp_neutral_tokens:
    if(final_tokens_1.count(word) and (neutral_tokens.count(word)==0)):
        neutral_tokens.append(word)

print(len(tmp_positive_tokens))
print(len(positive_tokens))

print(len(tmp_negative_tokens))
print(len(negative_tokens))

print(len(tmp_neutral_tokens))
print(len(neutral_tokens))


# print(positive_tokens)
```

```python
#taking input and checking sentiment

def checkSentiment(text1):
  text=text1.split()
  pos_value = 0
  neg_value = 0
  neu_value = 0
  for word in text:
      pos_value += positive_tokens.count(word)
      neg_value += negative_tokens.count(word)
      neu_value += neutral_tokens.count(word)
  print('pos: ' ,pos_value)
  print('neg: ' , neg_value)
  print('neu: ' , neu_value)
  if (pos_value > neg_value and pos_value > neu_value):
      print("it seems like a positive sentence")
  elif(pos_value==0 and neg_value==0 and neu_value == 0):
      print("it seems like a neutral vibe")
  elif(neg_value >= pos_value and neg_value >= neu_value):
      print("it seems like a negative sentence")
  else:
      print("it seems like a neutral vibe")

def checkSentiment1(text1):
  text= text1.split(' ')
  pos_value = 0
  neg_value = 0
  neu_value = 0
  for word in text:
      pos_value += positive_tokens.count(word)
      neg_value += negative_tokens.count(word)
      neu_value += neutral_tokens.count(word)
  if (pos_value > neg_value and pos_value > neu_value):
      return 'positive'
  elif(pos_value==0 and neg_value==0 and neu_value == 0):
      return 'neutral'
  elif(neg_value >= pos_value and neg_value >= neu_value):
      return 'negative'
  else:
      return 'neutral'
# text = input('enter your text!').split(' ')
checkSentiment(input('enter your text!   '))
```

```python
#testing
test_data = pd.read_csv('test(in).csv')
correct=0;
wrong=0;
i=0
while(i<len(test_data)):
  if(checkSentiment1(test_data['text'][i])==test_data['output'][i]):
    print('\033[1m' + 'COORRECT' + '\033[0m',i,': ',test_data['text'][i])
    print('predicted output: ',checkSentiment1(test_data['text'][i]))
    print('expected output: ',test_data['output'][i])
    correct += 1
  else:
    wrong += 1
    print('\033[1m' + 'WRONG' + '\033[0m',i,': ',test_data['text'][i])
    print('predicted output: ',checkSentiment1(test_data['text'][i]))
    print('expected output: ', test_data['output'][i])
  i+=1
print('total:',len(test_data))
print('correct:',correct)
print('wrong:',wrong)

print('accuracy: %.2f'%((correct/len(test_data))*100))
```
